```python
import pandas as pd
import lyricsgenius
import re
from retry import retry
import time

# Define your retry decorator
@retry(delay=5, backoff=2, max_delay=60, jitter=(0, 1), tries=3)
def fetch_lyrics(artist, song):
    # Search for the song lyrics
    search_query = f"{artist} {song}"
    song = genius.search_song(search_query)

    # Extract lyrics from the song object
    if song:
        lyrics = song.lyrics
        # Split lyrics by newline and exclude the first line (translation options)
        lyrics = '\n'.join(lyrics.split('\n')[1:])
        # Use regex to remove text enclosed in square brackets along with the brackets
        lyrics = re.sub(r'\[.*?\]', '  ', lyrics)
        # Replace "\n" with double spaces
        lyrics = lyrics.replace('\n', '  ')
        return lyrics
    else:
        return None

# Load your DataFrame
df = bigger_df # replace with your actual file name or DataFrame creation method

# Create a Genius API instance
genius = lyricsgenius.Genius("xj9277JxYnZE5LD6HUK9nb318EwRbIsYyXotDnRWt3R__BNaqhbuwehk61AaALyH")

# Initialize a list to store fetched lyrics
lyrics_list = []

# Record start time
start_time = time.time()

# Iterate through the DataFrame rows and fetch lyrics with retry mechanism
for index, row in df.iterrows():
    try:
        artist = row['artist_name']
        song = row['track_name']
        lyrics = fetch_lyrics(artist, song)
        if lyrics is None:
            lyrics = 'No result'  # Set "No result" for non-existent lyrics
        lyrics_list.append(lyrics)
    except Exception as e:
        lyrics_list.append('Error: {}'.format(str(e)))
    time.sleep(5)  # Add a delay between requests to avoid rate limiting
    
    # Check if 3 hours have passed, and if so, break out of the loop
    # if (time.time() - start_time) >= 3 * 60 * 60:
    #     print("3 hours elapsed. Stopping the code.")
    #     break

# # Add the lyrics list as a new column to the DataFrame
# df['lyrics'] = lyrics_list

# # Save the DataFrame to a new file
# df.to_csv("your_dataframe_with_lyrics.csv", index=False)  # replace with your desired file name
```
